```python
def find_frees(this_df):
    lengths = np.array([len(this_df["ball_z"][this_df["sec_in_quarter"]==i]) for i in this_df["sec_in_quarter"]])
    return(pd.DataFrame([lengths[np.where(lengths>5)[0]], this_df["sec_in_quarter"][np.where(lengths>5)[0]]], index=["Lengths", "Seconds Remaining"]).transpose().drop_duplicates())
```
